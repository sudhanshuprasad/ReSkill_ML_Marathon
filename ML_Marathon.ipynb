```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
def job_to_num(x):
        if x=='admin': return 12
        if x=='student': return 11
        if x=='unknown': return 10
        if x=='unemployed': return 9
        if x=='services': return 8
        if x=='self-employed': return 7
        if x=='retired': return 6
        if x=='entrepreneur': return 5
        if x=='blue-collar': return 4
        if x=='management': return 3
        if x=='technician': return 2
        if x=='housemaid':   return 1
        else: return 0
        
        
def marital_to_num(x):
    if x=='divorced': return 4
    if x=='married': return 3
    if x=='single': return 2
    if x=='unknown': return 1
    else: return 0


def education_to_num(x):
    if x=='tertiary': return 11
    if x=='secondary': return 10
    if x=='primary': return 9
    if x=='unknown': return 8
    if x=='university.degree': return 7
    if x=='professional.course': return 6
    if x=='illiterate': return 5
    if x=='high.school': return 4
    if x=='basic.9y': return 3
    if x=='basic.6y': return 2
    if x=='basic.4y': return 1
    else: return 0


def default_to_num(x):
    if x=='no': return 2
    if x=='yes': return 1
    else: return 0


def housing_to_num(x):
    if x=='no': return 2
    if x=='yes': return 1
    else: return 0

def loan_to_num(x):
    if x=='no': return 2
    if x=='yes': return 1
    else: return 0

def contact_to_num(x):
    if x=='cellular': return 2
    if x=='telephone': return 1
    else: return 0

def month_to_num(x):
    if x=='dec': return 12
    if x=='nov': return 11
    if x=='oct': return 10
    if x=='sep': return 9
    if x=='aug': return 8
    if x=='jul': return 7
    if x=='jun': return 6
    if x=='may': return 5
    if x=='apr': return 4
    if x=='mar': return 3
    if x=='feb': return 2
    if x=='jan': return 1
    else: return 0

def poutcome_to_num(x):
    if x=='failure': return 2
    if x=='success': return 1
    else: return 0

def deposit_to_num(x):
    if x=='no': return 2
    if x=='yes': return 1
    else: return 0
```

```python
df['job_num'] = df['job'].apply(job_to_num)
df['marital_num'] = df['marital'].apply(marital_to_num)
df['education_num'] = df['education'].apply(education_to_num)
df['default_num'] = df['default'].apply(default_to_num)
df['housing_num'] = df['housing'].apply(housing_to_num)
df['loan_num'] = df['loan'].apply(loan_to_num)
df['contact_num'] = df['contact'].apply(contact_to_num)
df['month_num'] = df['month'].apply(month_to_num)
df['poutcome_num'] = df['poutcome'].apply(poutcome_to_num)
df['deposit_num'] = df['deposit'].apply(deposit_to_num)
df
```

```python
df.drop(['job','marital','education','default','housing','loan','contact','month','poutcome','deposit'], axis=1, inplace=True)
#df.drop(['day'], axis=1, inplace=True)
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,16:].values
x,y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
y_train
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
lr.score(x_test, y_test)
print("Logistic Regression Test Accuracy {:.2f}%".format(lr.score(x_test, y_test)*100))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
print("Decision Tree Test Accuracy {:.2f}%".format(dtc.score(x_test, y_test)*100))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000, random_state = 1)
rf.fit(x_train, y_train.ravel())
my_prediction=rf.predict(x_test)
my_prediction
print("Random Forest Algorithm Accuracy Score : {:.2f}%".format(rf.score(x_test,y_test)*100))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train.ravel())
print("Naive Bayes Test Accuracy {:.2f}%".format(nb.score(x_test, y_test)*100))
```

```python
prediction_df = pd.DataFrame(my_prediction)
prediction_df
```

```python
compression_opts = dict(method='zip',
                        archive_name='out.csv')  
prediction_df.to_csv('prediction.zip', index=False,
          compression=compression_opts)  
```
